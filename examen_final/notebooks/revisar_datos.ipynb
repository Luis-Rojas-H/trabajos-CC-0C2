```python
import pandas as pd
import csv

col_names = [
    'id_oferta', 'id_usuario', 'target_elegibilidad', 
    'nom_facultad', 'nom_especialidad', 'cargo', 
    'funciones', 'conocimientos', 'experiencia', 
    'experiencia_cv', 'estudios_cv'
]

def leer_dataset(archivo):
    return pd.read_csv(
        archivo, 
        sep=',',               
        quotechar='"',         
        escapechar='\\',       
        engine='python',       
        on_bad_lines='warn',    
        encoding='utf-16',
        header=None,
        names=col_names
    )

df_con = leer_dataset('../datos/dataset_nlp_utf16_contratado.csv')
df_rech = leer_dataset('../datos/dataset_nlp_utf16_rechazado.csv')

df_final = pd.concat([df_con, df_rech], ignore_index=True)

df_final.to_csv(
    '../datos/dataset_nlp_utf16.csv', 
    index=False, 
    encoding='utf-16', 
    sep=',', 
    quoting=csv.QUOTE_ALL, 
    escapechar='\\'
)

print(f"Archivo unificado exitosamente: 'dataset_nlp_utf16.csv'")
print(f"Total de filas: {df_final.shape[0]}")
print(f"Total de columnas: {df_final.shape[1]}")

# Configuración de visualización para inspección final
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 10)

display(df_final)
```
