```python
import pandas as pd
import csv

col_names = [
    'id_oferta', 'id_usuario', 'target_elegibilidad', 
    'nom_facultad', 'nom_especialidad', 'cargo', 
    'funciones', 'conocimientos', 'experiencia', 
    'experiencia_cv', 'estudios_cv'
]

def leer_dataset(archivo):
    return pd.read_csv(
        archivo, 
        sep=',',               
        quotechar='"',         
        escapechar='\\',       
        engine='python',       
        on_bad_lines='warn',    
        encoding='utf-16',
        header=None,
        names=col_names
    )

df_con = leer_dataset('../datos/dataset_nlp_utf16_contratado.csv')
df_rech = leer_dataset('../datos/dataset_nlp_utf16_rechazado.csv')

df_final = pd.concat([df_con, df_rech], ignore_index=True)

df_final.to_csv(
    '../datos/dataset_nlp_utf16.csv', 
    index=False, 
    encoding='utf-16', 
    sep=',', 
    quoting=csv.QUOTE_ALL, 
    escapechar='\\'
)

print(f"Archivo unificado exitosamente: 'dataset_nlp_utf16.csv'")
print(f"Total de filas: {df_final.shape[0]}")
print(f"Total de columnas: {df_final.shape[1]}")

# Configuración de visualización para inspección final
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 10)

display(df_final)
```

```python
import re

def buscar_caracteres_extranos(df, patron):
    columnas_con_extranos = {}
    print("Buscando caracteres extraños.\n")

    for col in df.columns:
        if df[col].dtype == 'object':
            textos = df[col].dropna().astype(str)

            caracteres = textos.str.findall(patron).explode().dropna().unique()
            

            caracteres = [c for c in caracteres if c.strip() != '']

            if len(caracteres) > 0:
                columnas_con_extranos[col] = caracteres

    if columnas_con_extranos:
        print("Se encontraron caracteres extraños en las siguientes columnas:")
        for col, chars in columnas_con_extranos.items():
            print(f" -> Columna '{col}': {chars}")
    else:
        print("No se encontraron caracteres extraños en ninguna columna.")
```

```python
def inspeccionar_contexto(df, caracter, columnas, ventana=40, max_ejemplos=5):
    print(f"Buscando el carácter: {repr(caracter)}\n")

    encontrado_en_total = False

    for col in columnas:
        if df[col].dtype == 'object':
            textos = df[col].dropna().astype(str)

            mascara = textos.str.contains(caracter, regex=False)
            filas_con_caracter = textos[mascara]

            if not filas_con_caracter.empty:
                encontrado_en_total = True
                print(f"=== Columna '{col}' ({len(filas_con_caracter)} apariciones) ===")

                for idx, texto in filas_con_caracter.head(max_ejemplos).items():
                    pos = texto.find(caracter)

                    inicio = max(0, pos - ventana)
                    fin = min(len(texto), pos + len(caracter) + ventana)

                    fragmento = texto[inicio:fin]

                    fragmento_resaltado = fragmento.replace(caracter, f" >>>{caracter}<<< ")
                    
                    print(f" Fila {idx}: ...{fragmento_resaltado}...")
                print("-" * 60)
                
    if not encontrado_en_total:
        print(f"El carácter {repr(caracter)} no se encontró en las columnas indicadas.")
```

```python
# Diccionario con patrones Regex para destruir la estructura PHP sin tocar tu texto
diccionario_php = {
    r'a:\d+:\{': '',      # Borra el inicio del array (ej. a:1:{ o a:3:{)
    r'i:\d+;': '',        # Borra los índices de lista (ej. i:0;)
    r's:\d+:"': '',       # Borra el marcador de texto y tamaño (ej. s:393:")
    r'";\}': '',          # Borra el cierre final del array
    r'";': ' ',           # Borra el cierre entre elementos de la lista
    r'N;': ''             # Borra los valores nulos de PHP (N;)
}

print("Limpiando formato de PHP Serializado.\n")

columnas_php = ['funciones', 'conocimientos', 'experiencia']

for col in columnas_php:
    if df_final[col].dtype == 'object':
        for mal, bien in diccionario_php.items():
            # regex=True es obligatorio aquí porque estamos usando patrones dinámicos (\d+)
            df_final[col] = df_final[col].str.replace(mal, bien, regex=True)

print("Resultado limpio en la fila 6 de 'funciones':")
print(df_final.loc[6, 'funciones'])
```
